## File read

```python
## read file
file=open("git add.txt","r")
content=file.read()
print(type(content))
print(content)
```

```python
## in one line
file=open("git add.txt","r")
content = file.readlines()
print(content)
print("\n")
## line  by line print
with open("git add.txt","r") as file:
   for line in file:
      l=line.strip()
      print(l)
```

```python
## file close
file=open("git add.txt","r")
content=file.read()
file.close()
print(file.closed)

## another type 
with open("git add.txt","r") as file:
    content = file.readlines()
    print(content)
print(file.closed)
```

### File pointer

```python
## tell in file pointer
with open("test.txt","r") as file:
   print(file.tell()) ### Current file pointer/directory
   print(file.read())
   print(file.tell()) ### read full file
   print(file.read()) 
```

```python
with open("test.txt","r") as file:
   print(file.tell()) ### Current file pointer/directory
   print(file.read(11))
   print(file.tell()) ### read full file
   print(file.read()) 
```

```python
## seek in file pointer(left/right move)
with open("test.txt","r") as file:
   print(file.tell()) 
   file.seek(6)
   print(file.read(5))
   print(file.tell())
   file.seek(0)
   print(file.tell()) 
   print(file.read(5))
```
