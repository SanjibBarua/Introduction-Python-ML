### Tuples

```python
# declaration

tup=(10,20,30,40)
print(type(tup),tup)

#list to tuple

ls=[10,20,30,40]
tup1=tuple(ls)
print(type(tup1),tup1)

#access
n=(10,40,50,10.4,"str",False)
print(n[5],n[4])
```

```python
# tuple method

tup=(10,30,40,50,60,70,10,40,50,50,40)
print(tup.count(40))
print(tup.index(30))
```

## Methods of set

```python
a={1,3,5}
b={3,2,5,6,1}
print(a.union(b))
print(a.intersection(b))
print(a.isdisjoint(b))
print(a.issubset(b))
```

## Dictionary

```python
## declaration
## dic={key,value}
dic={"name" : "xyz", "age" : "20", "address" : "Bangladesh"}
print(type(dic),dic)

## access
print(dic["name"])
print(dic.get("age"))

dic["name"]="AI/ML"
print(dic.get("name"))

print(dic.get("number"))
print(dic.get("number",0))
```

```python
## adding value
dic={"name" : "xyz", "age" : "20", "address" : "Bangladesh"}
dic["number"]=30
print(dic)

dic.update({"rank":"Pro"})
print(dic)
```

```python
## deleting

del dic["number"]
print(dic)
```

```python
## view methods
dic={"name" : "xyz", "age" : "20", "address" : "Bangladesh"}
print(dic.keys())
dic["numbers"]=30
print(dic.keys())

print(dic.values())
print(dic.items())
```

```python
## iteration
for key,value in dic.items():
    print(key,value)
```

## Dictionary Comprehension

```python
square={x:x**2 for x in range(10)}
print(square)


co_ordinates=[(10,10.3),(20,20.3),(30,30.5)]
location=["Dhaka","Chattagram","Sylhet"]
exact_location={co_or : loc for co_or, loc in zip(co_ordinates,location)}
print(exact_location)
```
