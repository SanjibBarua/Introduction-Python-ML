### Tuples

```python
# declaration

tup=(10,20,30,40)
print(type(tup),tup)

#list to tuple

ls=[10,20,30,40]
tup1=tuple(ls)
print(type(tup1),tup1)

#access
n=(10,40,50,10.4,"str",False)
print(n[5],n[4])
```

```python
# tuple method

tup=(10,30,40,50,60,70,10,40,50,50,40)
print(tup.count(40))
print(tup.index(30))
```
