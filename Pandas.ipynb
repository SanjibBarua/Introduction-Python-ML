## Changing Index and Columns and iloc

```python
df_index=df.set_index('StudentID')
df_index
```

```python
df_index.iloc[3:7]
```

```python
df.rename(columns={'FullName':'Full Name'},inplace=True)
df
```

```python
## row delete
df.drop(19,inplace=True)
df
```

```python
# column delete
df.drop('Location',axis=1,inplace=True)
df
```

```python
## value asign 
df.loc[1,'Python Marks']=80
df.loc[1,'CompletionStatus']='Completed'
df
```

```python
##Iteration

for i,series in df.iterrows():
    print(f"{i} : {series}")
```
