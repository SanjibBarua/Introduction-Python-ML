### String methods

```python
string="Welcome to Ai and ml course,'ai',hello ai,to ai"

processed_string=string.upper()
print(processed_string)

processed_string=string.lower()
print(processed_string)

# length
print(len(string))

# check
print("Ai and ml" in string)

#  first substring 
f_index=processed_string.find("ai and ml")
print(f_index)

# last substring
l_index=processed_string.rfind("ai")
print(l_index)

#substring counting
counting=processed_string.count("ai")
print(counting)

#replace a substring
new_string=string.replace("ai","ml")
print(new_string)
```

## Splitting,Joining and Formating

```python
prompt="what is Python ?"
tokens=prompt.split()
print(tokens)

## anyother 
prompt="what-is-Python-?"
tokens=prompt.split("-")
print(tokens)
```

```python
new = ",".join(tokens)
print(new)
```

## List & Methods

```python
## declaration

number=[10,20,30,40]
print(type(number))
```

```python
## access

print(number[2])

number[2]=90
print(number[2])
```

```python
new_list=number[0:3]
print(new_list)
```

```python
## modification

number=[10,20,30,40]
## add value
number.append(80)
print(number)

## insertion in any position
number.insert(0,60)
print(number)

## deleting
number.pop()
print(number)

```

```python
number=[10,20,30,40]
number.sort()
print(number)
number.sort(reverse=True)
print(number)
```

### list as stack

```python
stack=[]
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
stack.append(5)


## top element 
print(stack[-1])
print(" ")
## access element
print(stack[4])
print(" ")
##getting top value
print(stack.pop())
print(stack.pop())
print(stack.pop())
print(stack.pop())
print(stack.pop())
```

### list as queue

```python
queue=[]
queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)

## remove elements
print(queue.pop(0))

```

### List Comprehension

```python
random_list=[x for x in range(20) if x%2==0]
print(random_list)
```

```python
### List Comprehension in string

fruits=['apple','mango','orange','banana']
upper_fruits=[fruits.upper() for fruits in fruits]
print(upper_fruits)
```
