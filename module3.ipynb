### String methods

```python
string="Welcome to Ai and ml course,'ai',hello ai,to ai"

processed_string=string.upper()
print(processed_string)

processed_string=string.lower()
print(processed_string)

# length
print(len(string))

# check
print("Ai and ml" in string)

#  first substring 
f_index=processed_string.find("ai and ml")
print(f_index)

# last substring
l_index=processed_string.rfind("ai")
print(l_index)

#substring counting
counting=processed_string.count("ai")
print(counting)

#replace a substring
new_string=string.replace("ai","ml")
print(new_string)
```
